```python
import pandas as pd
import numpy as np
import cv2
import os
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
df['width'] = 0
df['height'] = 0
df
```

```python
DIR = 'Medical mask/images/'

for idx, row in df.iterrows():
    filename = os.path.join(DIR, row['name'])
    height, width = cv2.imread(filename).shape[:2]
    df.loc[idx,'width'] = width
    df.loc[idx,'height'] = height
```

```python
df
```

```python
df_maks = pd.read_csv('train_maks.csv')
df_maks
```

```python
df_maks = df_maks.drop(['Unnamed: 0'], axis=1)
df_maks
```

```python
df_all = pd.concat([df, df_maks], ignore_index=True)
df_all
```

```python
df_all.to_csv('train_all.csv', index=False)
```
