```python
import pandas as pd
import numpy as np
import os

import cv2
from tqdm import tqdm
```

```python
df = pd.DataFrame(columns=['name','x1','y1','x2','y2','classname','width','height'])
df
```

```python
f = open('images/wider-train.list')

contents = f.readlines()

for content in tqdm(contents):
    content = content.strip()
    image_path = content
    path = content.split('/')
    label_path = 'labels/'+ '/'.join(path[1:])
    label_path = label_path.split('.')[0] + '.txt'
    
    name = path[-1]
    height, width = cv2.imread(image_path).shape[:2]
    
    try:
        fm = open(label_path)
    except FileNotFoundError:
        print('Label 파일이 없습니다!')
    else:
        bboxs = fm.readlines()
        for bbox in bboxs:
            bbox = bbox.strip()
            # 0 for face_no_mask 1 for face_with_mask
            is_mask, x0_rel, y0_rel, w_rel, h_rel = bbox.split(' ')

            if is_mask == '0':
                classname = 'face_no_mask'
            else:
                classname = 'face_with_mask'

            xc = int(width*float(x0_rel))
            yc = int(height*float(y0_rel))
            bw = int(width*float(w_rel))
            bh = int(height*float(h_rel))
            x1 = xc-(bw//2)
            y1 = yc-(bh//2)
            x2 = xc+(bw//2)
            y2 = yc+(bh//2)

            row = {'name': name, 'x1':x1, 'y1': y1, 'x2':x2, 'y2':y2,
                   'classname': classname, 'width': width, 'height': height}

            df = df.append(row, ignore_index=True)
        fm.close()
f.close()
```
